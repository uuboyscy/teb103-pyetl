```python
import jieba
import os
import glob
```

```python
sb = ''
for f in glob.glob('./res_gossiping/*.txt'):
    with open(f, 'r', encoding='utf-8') as file:
        sb += file.read().split('---split---')[0]
print(sb)
```

```python
jieba.load_userdict('./mydict.txt')
sbCut = jieba.cut(sb)
```

```python
# for i in sbCut:
#     print(i)
sbCutList = list()
for w in sbCut:
    if len(w) > 1:
        sbCutList.append(w)

sbCutList = [w for w in sbCut if len(w) > 1]
sbCutList
```

```python
wordCount = dict()
for w in sbCutList:
    if w in wordCount:
        wordCount[w] += 1
    else:
        wordCount[w] = 1

wordCount
```

```python
sortList = list()
for w in wordCount:
    sortList.append([w, wordCount[w]])
sortList
```

```python
sortList.sort(key=lambda x: x[1], reverse=True)
sortList
```
