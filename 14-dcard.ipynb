## 完整程式

```python
import requests
import json
from urllib import request
import os
import ssl
ssl._create_default_https_context=ssl._create_unverified_context

headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 11_0_0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36'
}
url = 'https://www.dcard.tw/service/api/v2/forums/photography/posts?limit=30&before=234750076'

res = requests.get(url, headers=headers)

jsonStr = res.text # string

jsonData = json.loads(jsonStr)

if not os.path.exists('./dcard'):
    os.mkdir('./dcard')

for articleDict in jsonData:
    title = articleDict['title']
    articleUrl = 'https://www.dcard.tw/f/photography/p/' + str(articleDict['id'])
    imgList = articleDict['mediaMeta'] # list
    
    print(title)
    print(articleUrl)
    
    for imgUrl in imgList:
        print('\t' + imgUrl['url'], end='')
        imgPath = './dcard/' + imgUrl['url'].split('/')[-1]
        try:
#             request.urlretrieve(imgUrl['url'], imgPath)
            resImg = requests.get(imgUrl['url'], headers=headers)
            with open(imgPath, 'wb') as f:
                f.write(resImg.content)
            print(' done')
        except Exception as e:
            print(e)
            print(' error')
```
