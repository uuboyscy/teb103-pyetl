```python
import requests
from bs4 import BeautifulSoup

userAgent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 11_0_0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.193 Safari/537.36'
headers = {
    'User-Agent': userAgent
}

url_landingpage = 'https://www.ptt.cc/ask/over18?from=%2Fbbs%2FGossiping%2Findex.html'
url_tmp = 'https://www.ptt.cc/ask/over18'
url = 'https://www.ptt.cc/bbs/Gossiping/index.html'

ss = requests.session()

print(ss.cookies)

res = ss.get(url_landingpage, headers=headers)

print(ss.cookies)

soup = BeautifulSoup(res.text, 'html.parser')

data = dict()

key1 = soup.select_one('input')['name']
value1 = soup.select_one('input')['value']
data[key1] = value1

key2 = soup.select('button')[0]['name']
value2 = soup.select('button')[0]['value']
data[key2] = value2
# print(data)

ss.post(url_tmp, data=data)

print(ss.cookies)

res = ss.get(url, headers=headers)
print(res.text)
```
