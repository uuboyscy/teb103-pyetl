```python
import pandas as pd
import glob

data = list()
for f in glob.glob('./res_gossiping/*.txt'):
    with open(f, 'r', encoding='utf-8') as file:
        data.append(file.read().split('---split---')[1].split('\n')[1:-1])
        
data
```

```python
columns = [s.split(':')[0] for s in data[0]]
columns
```

```python
df = pd.DataFrame(data=data, columns=columns)
df
```

```python
def columnFilter(s):
    output = s.split(': ')[1]
    return output
```

```python
df2 = df.copy()
```
